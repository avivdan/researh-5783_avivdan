```python
def last_call(func):
    '''
    >>> pow2(x=2)
    4
    >>> pow2(x=2)
    "dealing with dementia?, *('x', 2)* was already checked"
    >>> pow2(2)
    4
    >>> pow2(2)
    'dealing with dementia?, *2* was already checked'
    >>> hey(x="wow")
    'hey : wow'
    >>> hey("wow")
    'hey : wow'
    >>> hey(x="wow")
    "dealing with dementia?, *('x', 'wow')* was already checked"
    '''
    saved = []
    def wrapper(*args , **kwargs):
        for arg in args:
            if arg in saved:
                return f"dealing with dementia?, *{arg}* was already checked"
            else:
                saved.append(arg)
                return func(*args, **kwargs)
        for kw in kwargs.items():
            if kw in saved:
                return f"dealing with dementia?, *{kw}* was already checked"
            else:
                saved.append(kw)
                return func(*args, **kwargs)
    return wrapper

@last_call
def pow2(x:int):
    return x**2

@last_call
def div2(x:int):
    return x/2

@last_call
def pow2(x):
    return x**2

@last_call
def div2(x):
    return x/2

@last_call
def hey(x:str):
    return "hey : "+x

if __name__ == "__main__":
    import doctest
    doctest.testmod(verbose=True)
```
