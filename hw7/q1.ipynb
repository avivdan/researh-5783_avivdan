```python
import cvxpy as cp
import numpy as np
import time

# Problem data.
lst_np = []
lst_cp = [] 
for i in range(5):
    np.random.seed(1)
    m = 200*(3*i+1)
    n = m
    A = np.random.randn(m, n)
    b = np.random.randn(m)   
    start_time = time.time()
    solution = np.linalg.solve(A, b)
    lst_np.append((time.time() - start_time))
    start_time = time.time()
    x = cp.Variable(n)
    objective = cp.Minimize(cp.sum(A@x - b))
    constraints = [A@x - b == 0]
    prob = cp.Problem(objective, constraints)
    prob.solve()
    lst_cp.append((time.time() - start_time))
```

```python
import matplotlib.pyplot as plt

sec = list(i for i in range(5))
values_cp = list(lst_cp)
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(sec, values_cp, color ='maroon',
        width = 0.4, )
 
plt.ylabel("time")
plt.xlabel("run number")
plt.title("cvxpy")
plt.show()
```

```python
sec = list(i for i in range(5))
values_np = list(lst_np)
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(sec, values_np, color ='maroon',
        width = 0.4, )
 
plt.ylabel("time")
plt.xlabel("run number")
plt.title("numpy")
plt.show()
```

In conclusion numpy solves much faster than cvxpy linear equation

```python
ploting = [(lst_cp[i]/lst_np[i]) for i in range(5)]
ploting

sec = list(i for i in range(5))
values = list(ploting)
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(sec, values, color ='maroon',
        width = 0.4, )
 
plt.ylabel("ratio")
plt.xlabel("run number")
plt.title("ratio")
plt.show()
```
